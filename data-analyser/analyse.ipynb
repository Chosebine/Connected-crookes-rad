```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as fft   
```

```python
df = pd.read_hdf('~/Connected-crookes-rad/data.h5')
df = df.iloc[2:,:]
```

```python
plt.figure(figsize=(15,5),dpi=300)
fig=plt.plot(df.date,df.varlux,linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(15,5),dpi=300)
fig=plt.plot(df.date,df.lux.rolling(10).mean(),linewidth=0.5)
plt.show()
```
